## Visualize results

#### Import libraries

```python
from pyproj import CRS
import geopandas as gpd
from functools import reduce
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import contextily as ctx
```

#### Read population, rails, subway stations and rail stations data

```python
# Read stations population data
fp = "population_data/stations_population.shp"
stations_population = gpd.read_file(fp)

stations = list(stations_population["Name"].unique())

stations_population.head(3)
```

#### Create and save annual images

```python
# Create and save figure of every year in the seleted list

# Select colour map
#cmap = matplotlib.cm.autumn
cmap = matplotlib.cm.get_cmap('autumn_r')
#cmap = matplotlib.cm.gnuplot
#cmap = matplotlib.cm.get_cmap('gnuplot_r')

# Get columns names
columns = stations_population.columns.tolist()

# Filter columns list and leave only population columns (they have year eg. pop1997)
pop_years = list(filter(lambda x: 'pop' in x, columns))

# Remove pop and leave only years. Then years list is ready for loop.
years_list=[]
for name in pop_years:
    year = name[3:7]
    year = int(year)
    years_list.append(year)

for station in stations:
    name = station
    data = stations_population[stations_population["Name"] == station]
    #population = data[pop_years]
    population = []
    for year in pop_years:
        population.append(int(data[year]))
    
    #print(name) # Title
    #print(years_list) # X
    #print(population) # Y
    
    # plotting the points  
    plt.plot(years_list, population) 
    # naming the x axis 
    plt.xlabel('Year') 
    # naming the y axis 
    plt.ylabel('Population') 
    # giving a title to my graph 
    plt.title(name) 
    # Save fig
    plt.savefig("pics_stations_graphs/"+name+".png", dpi=100)
    # function to show the plot 
    plt.show() 
```
